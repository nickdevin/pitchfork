# **Detecting Bias in Music Reviews at pitchfork.com**

With nearly a quarter of a million readers per day, the music review publication Pitchfork is a hugely influential force in the independent music scene. A positive album review from Pitchfork can help launch an artist into more mainstream success, notable examples of which include Arcade Fire and Sufjan Stevens.

Pitchfork has also been the subject of criticism in regards to certain perceived biases in their content. A review of hip-hop artist M.I.A.'s 2007 album *Kala* erroneously stated that Diplo had produced all of the album's tracks, when, in reality, M.I.A. had produced most of the tracks herself. In an interview with Pitchfork, M.I.A. later said, "There is an issue especially with what male journalists write about me and say 'this *must* have come from a guy.'" Their early reviews often focused disproportionately on independent rock, though they appear in recent years to have featured a more balanced mixture of genres.

My goal in this project is to examine bias in the bias in the reviews, specifically looking to answer the following questions:
- How has the proportion of reviews per genre evolved over time?
- Is there evidence of bias in scoring based on the genre of an album?
- Is there evidence of gender bias in Pitchfork's review? Specifically:
    - How has the ratio of reviews of female artists to male artists evolved over time?
    - Is there evidence of bias in scoring based on the artist's gender or on the reviewer's gender?

## Genre Analysis

```python
# Restrict to reviews for which genre is not missing

genre_df = pitchfork.copy()
genre_df = genre_df[~genre_df['genre'].isnull()]
genre_df = genre_df[['genre', 'score', 'review_date']]
genre_df
```

```python
# Several albums belong to more than one genre.
# This splits a single observation with multiple genres into a separate observation, one for each genre.

genres_split = genre_df['genre'].map(lambda s: s.split('|'))
genres_flattened = []
for index in genres_split.index:
    genres_flattened.extend([[index, genre] for genre in genres_split[index]])
to_join = pd.DataFrame(genres_flattened, columns = ['index', 'genre'])
genre_split_df = pd.merge(genre_df.drop('genre', axis = 1), to_join, right_on = 'index', left_index = True)
genre_split_df.head(10)
```

```python
# The following list will be useful later.

unique_genres = genre_split_df['genre'].unique().tolist()
unique_genres
```

### What is the trend of mean review score by genre per year?

```python
genre_counts = genre_split_df.groupby(['review_date', 'genre']).count()
genre_counts = genre_counts.reset_index()
genre_counts = genre_counts[['review_date', 'genre', 'score']]
genre_counts.columns = genre_counts.columns.str.replace('score', 'count')
genre_counts.head(10)
```

```python
# Check the number of reviews per year.

year_df = pitchfork.copy()
year_df[['review_date', 'score']].groupby('review_date').count()
```

```python
genre_facet = sns.FacetGrid(data = genre_counts, col = 'review_date', col_wrap = 3, height = 4, sharey = False, hue = 'genre', palette = 'rocket')
genre_graphs = genre_facet.map(sns.barplot, 'genre', 'count', order = unique_genres)
for axis in genre_graphs.axes.flatten():
    axis.tick_params(labelbottom=True)
genre_graphs.set_xticklabels(rotation=90)
plt.subplots_adjust(hspace=1, wspace=0.5)
plt.title('Number of Reviews per Genre')
```

### Conclusions drawn from graphs

In each year, rock is by far the most reviewed genre, and with electronic music the second-most reviewed genre in all years except 2018. There are some noteworthy changes in the distributions over time, with relatively more reviews for pop/R&B and rap albums beginning around 2015. The distributions from 2012 onward are more balanced, overall. Between 2003 and 2011, there were approximately 600 reviews for albums classified as "rock", and this number dropped steadily to approximately 400 per year, while the total number of reviews did not change significantly.

### Is there a significant difference in average score by genre?

Before running a one-way ANOVA test, note that, since an album can be classified as several genres, the occurrences of different genres may not be independent of each other: for instance, a rock album may be more likely also to be classified as "metal" than as "jazz."

We run a $\chi^2$ test to assess the independence of the genres.

```python
genre_series = pitchfork['genre'].copy()
chi_columns = [genre_series.dropna().map(lambda t: s in str(t)) for s in unique_genres]
chi_df = pd.DataFrame(np.matrix(chi_columns).T, columns = unique_genres)
chi_df.head(10)
```

```python
chi_counts = chi_df.groupby(unique_genres).apply(lambda group: group.shape[0])
chi_counts
```

```python
def genre_combination_count(x):
    try:
        count = chi_counts[x]
    except:
        count = 0
    return count
```

```python
combination_list = list(product([True, False], repeat=len(unique_genres)))
chi_array = np.array(list(map(genre_combination_count, combination_list))).reshape(*[2]*len(unique_genres))
chi_array
```

```python
chi2statistic, P, df, expected = stats.chi2_contingency(chi_array)
print(P)
```

The P-value is ≈ 0, so the occurrence of one genre is NOT necessary independent of another.
We need to exercise more caution when applying one-way ANOVA, as one genre's average score is not necessarily independent of another's.
To remedy this, we will see if there is a difference between average scores of albums belonging to single genres only.

```python
single_genre_df = pitchfork[['genre', 'review_date', 'score']].copy()
mask = single_genre_df['genre'].map(lambda genre: genre in unique_genres)
single_genre_df = single_genre_df[mask]
```

```python
genre_means = single_genre_df.groupby(['review_date', 'genre'])['score'].mean().reset_index()
```

```python
genre_means
```

```python
genre_means_facet = sns.FacetGrid(data = genre_means, col = 'review_date', col_wrap = 3, height = 4, hue = 'genre', palette = 'rocket')
genre_means_graphs = genre_means_facet.map(sns.barplot, 'genre', 'score', order = unique_genres)
for axis in genre_means_graphs.axes.flatten():
    axis.tick_params(labelbottom=True)
genre_means_graphs.set_xticklabels(rotation=90)
plt.subplots_adjust(hspace=1, wspace=0.5)
plt.title('Average Review Score by Genre')
```

```python
# one-way ANOVA for average review score overall
genres_separate = [single_genre_df[single_genre_df['genre'] == genre]['score'] for genre in unique_genres]
stats.f_oneway(*genres_separate)

# There IS a significant difference between average scores between genres.
```

```python
# one-way ANOVA each year in which at least one album per genre was reviewed (2001-2020):
years_separate = [single_genre_df[single_genre_df['review_date'] == year] for year in range(2001, 2021)]
```
